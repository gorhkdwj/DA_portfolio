### 문제 5

다음 리스트에서 짝수만 골라 새로운 리스트를 만들어 출력하세요.

`nums = [3, 10, 7, 8, 2, 9, 12]`

````python
# for 문을 활용한 조건 반복문

nums=[3, 10, 7, 8, 2, 9, 12]
new_list=[]
for i in nums:
  if i%2==0:
    new_list.append(i)
print(new_list)
````

````python
# List Comprehension을 활용
# if 조건 없이도, List Comprehension이 가능하다...!!
# 간결하고 효율적인 방법으로 새로운 리스트를 생성하는 방법일 뿐! 그저 활용도가 높은 친구였을 뿐이다.

nums=[3, 10, 7, 8, 2, 9, 12]
new_list=[i for i in nums if i%2==0]
test=[i for i in nums]
print(new_list)
print(test)
````

### 문제 6

```python

```

````python
# a= [1,2,3]
# b= [1,2,3,4]
````

````python
a = [1, 2, 3]
print(id(a))
b = a
b.append(4)
print(id(b))
print(a)
print(b)
# 리스트를 생성하면 해당 리스트의 메모리 주소값이 고정된다.
# 변수는 해당 메모리에 저장된 객체에 이름표를 붙이는 것일 뿐!
# 따라서 append(4)를 통해 b라는 이름표가 붙은 [1,2,3]을 [1,2,3,4]로 변경한다면
# 메모리 id는 고정된 채 객체만 바뀌게 된다.
# 고정된 메모리 id에 붙은 이름표 a, b 무엇을 부르더라도 같은 객체를 지칭하며 해당 개체가 변경된 값을 똑같이 가져가게 된다.
# 단!! a=[1,2,3], b=[1,2,3]을 각각 만들게 되면 리스트는 가변 객채집합이기 때문에 b에 새로운 메모리 id가 부여된다.
x=[1,2,3,4,5]
y=[1,2,3,4,5]
print(id(x))
print(id(y))
# 위와 같이 다른 id를 갖는다.
````

### 문제 9

다음 코드의 출력 결과를 예측하세요.

```python
d = {"a": 1, "b": 2}
print(d.get("c"))
print(d.get("c", 0))

```
이 문제에서 d.get("c",0)은 다음과 같은 의미를 갖는다고 한다!(Gemini)

d.get("c", 0)의 의미: "보험 들기"
이 코드는 딕셔너리에게 이렇게 부탁하는 것입니다.

"혹시 'c'라는 키 있어? 있으면 그 값을 주고, 없으면 에러 내지 말고 그냥 0을 줘."

"c": 찾고 싶은 열쇠(Key)

0: 열쇠가 없을 때 대신 내놓을 기본값(Default Value)

````python
# none과 0을 각각 출력하게 된다.
# print(d.get("c")) => value 없음! => None!
# print(d.get("c", 0)) => value 없음! -> (없을때 0 달라고?) => 0!
d = {"a": 1, "b": 2}
print(d.get("c"))
print(d.get("c", 0))
print(d)
````

### 문제 10

```python
d = {"x": [1, 2], "y": [3, 4]}

print(d)

```
