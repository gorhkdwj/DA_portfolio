### 1. 문제 설명

> 정수 num1과 num2가 주어질 때, num1에서 num2를 뺀 값을 return하도록 solution 함수를 완성해주세요.

### 2. 제한사항

* -50000 ≤ num1 ≤ 50000

### 3. 입출력 예

| num1 | num2 | result |
| :--- | :--- | :--- |
| 2 | 3 | -1 |
| 100 | 2 | 98 |

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: (이 문제를 풀기 위해 필요한 핵심 아이디어를 한 줄로 적어두세요. 예: 리스트 컴프리헨션 사용, 스택 구조 활용 등)

```python
# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. 뺄셈 연산자(-)를 사용한다.
# 3. 결과를 반환한다.

def solution(num1, num2):
    answer = num1 - num2
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(2, 3))   # 예상: -1
print(solution(100, 2)) # 예상: 98
```

def와 return 함수는 처음 봐서 당황스러웠지만, solution이라는 함수를 생성하여 return을 통해 값을 반환한다 라는 개념을 알게 되었다.
