### 1. 문제 설명

> 정수 num1, num2가 매개변수 주어집니다. num1과 num2를 곱한 값을 return 하도록 solution 함수를 완성해주세요.

* 0 ≤ num2 ≤ 100

### 3. 입출력 예

입출력 예 #1
* num1이 3, num2가 4이므로 3 * 4 = 12를 return합니다.

```python
# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. 곱셈 연산자(*)를 사용한다.
# 3. 결과를 반환한다.

def solution(num1, num2):
    answer = num1*num2
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(2, 3))   # 예상: 6
print(solution(100, 2)) # 예상: 200
```

### 1. 문제 설명

> 정수 num1, num2가 매개변수로 주어질 때, num1을 num2로 나눈 몫을 return 하도록 solution 함수를 완성해주세요.

* 0 < num1 ≤ 100
* 0 < num2 ≤ 100

### 3. 입출력 예

입출력 예 #1
* num1이 10, num2가 5이므로 10을 5로 나눈 몫 2를 return 합니다.
입출력 예 #2

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: (이 문제를 풀기 위해 필요한 핵심 아이디어를 한 줄로 적어두세요. 예: 리스트 컴프리헨션 사용, 스택 구조 활용 등)

```python
# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. 나눗셈 연산자(//)를 사용한다.
# 3. 결과를 반환한다.

def solution(num1, num2):
    answer = num1//num2
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(7, 3))   # 예상: 2
print(solution(100, 2)) # 예상: 50
```

# [Lv.1] 문제 제목 (나이 출력)

- **출처**: 프로그래머스
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/120820)
- **카테고리**: 연산

---

### 1. 문제 설명

> 머쓱이는 선생님이 몇 년도에 태어났는지 궁금해졌습니다. 2022년 기준 선생님의 나이 age가 주어질 때, 선생님의 출생 연도를 return 하는 solution 함수를 완성해주세요

### 2. 제한사항

* 0 < age ≤ 120
* 나이는 태어난 연도에 1살이며 매년 1월 1일마다 1살씩 증가합니다.


### 3. 입출력 예

입출력 예 #1
* 2022년 기준 40살이므로 1983년생입니다.
입출력 예 #2
* 2022년 기준 23살이므로 2000년생입니다.

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: 제한사항에 따라 나이를 구하기 위해서는 [기준연도-나이+1] 로 계산해야 한다.
---


```python
# [풀이 전략]
# 1. 나이를 입력받는다.
# 2. 기준연도에서 나이를 빼고 1을 더한다
# 3. 결과를 반환한다.

# 기준: 2022년
def solution(age):
    answer = 2022-age+1
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(32))   # 예상: 1991
print(solution(50)) # 예상: 1973
```

# [Lv.1] 문제 제목 (숫자 비교하기)

- **출처**: 프로그래머스
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/120807)
- **카테고리**: if 조건문

---

### 1. 문제 설명

> 정수 num1과 num2가 매개변수로 주어집니다. 두 수가 같으면 1 다르면 -1을 retrun하도록 solution 함수를 완성해주세요.

### 2. 제한사항

* 0 ≤ num1 ≤ 10,000
* 0 ≤ num2 ≤ 10,000

### 3. 입출력 예

입출력 예 설명 #1

num1이 2이고 num2가 3이므로 다릅니다. 따라서 -1을 return합니다.
입출력 예 설명 #2

num1이 11이고 num2가 11이므로 같습니다. 따라서 1을 return합니다.
입출력 예 설명 #3

num1이 7이고 num2가 99이므로 다릅니다. 따라서 -1을 return합니다.

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: 두 수가 같은 경우와 다른 경우를 나눠야 하기 때문에 if문을 활용한다!
---


```python
# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. if 구문을 사용해 두 수가 같은 경우와 다른 경우를 나눈다.
# 3. 각 경우에 반환할 값을 answer 변수에 지정한다.
# 4. 결과를 반환한다.

def solution(num1, num2):
    if num1==num2:
        answer=1
    else :
        answer=-1
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(1, 1))   # 예상: 1
print(solution(10, 5)) # 예상: -1
```

# [Lv.1] 문제 제목 (두 수의 나눗셈)

- **출처**: 프로그래머스
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/120806)
- **카테고리**: 연산

---

### 1. 문제 설명

> 정수 num1과 num2가 매개변수로 주어질 때, num1을 num2로 나눈 값에 1,000을 곱한 후 정수 부분을 return 하도록 solution 함수를 완성해주세요.

### 2. 제한사항

* 0 < num1 ≤ 100
* 0 < num2 ≤ 100

### 3. 입출력 예

입출력 예 #1

num1이 3, num2가 2이므로 3 / 2 = 1.5에 1,000을 곱하면 1500이 됩니다.
입출력 예 #2

num1이 7, num2가 3이므로 7 / 3 = 2.33333...에 1,000을 곱하면 2333.3333.... 이 되며, 정수 부분은 2333입니다.
입출력 예 #3

num1이 1, num2가 16이므로 1 / 16 = 0.0625에 1,000을 곱하면 62.5가 되며, 정수 부분은 62입니다.

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: / 연산자로 나눠준 후 정수형 변환을 통해 정수 부분만 반환하도록 한다!
---


```python
# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. 나눗셈 연산자(/)를 사용하여 두 수를 나누고 곱셈 연산자(*)를 사용하여 1000을 곱한다.
# 3. 결과를 정수형으로 변환한다
# 4. 결과를 반환한다.

def solution(num1, num2):
    answer = num1/num2*1000
    answer = int(answer)
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(3, 2))   # 예상: 1500
print(solution(100, 2)) # 예상: 50000
```
