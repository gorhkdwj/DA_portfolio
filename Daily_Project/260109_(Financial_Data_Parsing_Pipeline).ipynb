Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
상황: 새로운 데이터 {"ticker": "SPY", "ror": 11.2}와 {"ticker": "NVDA", "ror": 25.0}이 들어왔습니다.

미션:

SPY는 이미 있으니 수익률을 11.2로 업데이트해야 합니다.

NVDA는 없으니 리스트에 **추가(append)**해야 합니다.

이를 처리하는 for 반복문 로직을 작성하세요. (힌트: found 플래그 변수 활용)

```python
# 추가할 포트폴리오
while True: # [외부 반복문] 전체 프로세스 반복
    try:
        target_ticker=input("추가할 티커를 입력하세요.(종료는 'exit' 입력): ").upper() # 티커 입력
    except ValueError: # 입력 타입 오류
        print("티커는 문자로 입력해주세요.")
        continue
    if target_ticker=='EXIT': # 종료 조건
        print("입력을 종료합니다.")
        break
    existing_tickers=[item['ticker'] for item in portfolio] # 이미 구성된 티커 모음
    if target_ticker in existing_tickers: # 구성된 티커와 중복 체크
        print("이미 추가된 티커입니다.")
        continue
    while True:# [내부 반복문] 미 존재 시 수익률 입력값이 잘못 되면 티커부터 새로 입력해야 함.
        try:
            target_ror= float(input("추가할 수익률을 입력하세요.: ")) # 수익률 입력(실수형)
            # 수익률이 숫자인 경우 내부 반복문 break 후 티커입력부터 외부 반복문 시행
            break
        except ValueError: # 수익률 입력이 숫자가 아닌 경우 내부 반복문 시행 후 수익률 재입력
            print("수익률은 숫자로 입력해주세요.")
    portfolio.append({"ticker": target_ticker, "ror": target_ror}) # 포트폴리오에 추가
    print(f"({target_ticker}: RoR={target_ror})(이)가 추가되었습니다.")
print(f"추가 완료된 포트폴리오.: ")
for item in portfolio:
    print(f"{item['ticker']}: RoR={item['ror']}")
```

```python
# 수정할 티커와 수익률
while True:
    target_ticker= input("수정할 티커를 입력하세요.(종료는 'exit' 입력): ").upper()
    if target_ticker=='EXIT':
        print("수정이 완료되었습니다.")
        break
    for item in portfolio:
        if item['ticker']==target_ticker:
            item['ror']=float(input("수정할 수익률을 입력하세요."))
            print(f"수정한 {target_ticker}의 수익률은 {item['ror']}입니다.")
            break
    else:
        print("해당 티커가 없습니다.")
        break
print(f"수정된 포트폴리오: ")
for item in portfolio:
    print(f"{item['ticker']}: {item['ror']}")
    
            
        
```
