Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
상황: 새로운 데이터 {"ticker": "SPY", "ror": 11.2}와 {"ticker": "NVDA", "ror": 25.0}이 들어왔습니다.

미션:

SPY는 이미 있으니 수익률을 11.2로 업데이트해야 합니다.

NVDA는 없으니 리스트에 **추가(append)**해야 합니다.

이를 처리하는 for 반복문 로직을 작성하세요. (힌트: found 플래그 변수 활용)

```python
# 추가할 티커와 수익률을 입력하세요. (예: TICK, ror)
# tick= input("추가할 티커를 입력하세요.: ")
# ror= float(input("추가할 수익률을 입력하세요.: "))
# portfolio.append({"ticker": tick, "ror": ror})

while True:
    tick= input("추가할 티커를 입력하세요.(종료는 'exit' 입력): ")
    if tick=='exit':
        break
    ror= float(input("추가할 수익률을 입력하세요.: "))
    portfolio.append({"ticker": tick, "ror": ror})
print(portfolio)
```
