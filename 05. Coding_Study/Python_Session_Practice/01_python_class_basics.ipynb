### 1. 왜 클래스가 필요한가요? (불편한 상황 체험)

여러분이 선생님이 되어 학생 3명의 성적을 관리한다고 가정해 봅시다.
클래스를 모른다면, 우리는 학생 한 명 한 명을 위해 변수를 계속 만들어야 합니다.

```python
# [상황] 성적 처리 시스템을 만드는데, 학생이 늘어날 때마다 변수 이름 짓기가 고통스럽습니다.

# 학생 1
student1_name = "철수"
student1_score = 90

# 학생 2
student2_name = "영희"
student2_score = 85

# 학생 3
student3_name = "민수"
student3_score = 95

# 출력하기도 힘듭니다. (변수명을 다 기억해야 함)
print(f"{student1_name} : {student1_score}점")
print(f"{student2_name} : {student2_score}점")
print(f"{student3_name} : {student3_score}점")
```

> **문제점:** 
> 만약 학생이 100명이라면? `student100_name`까지 변수 200개를 만들어야 합니다.
> 오타가 날 확률도 매우 높고, 관리가 불가능에 가깝습니다.

---

### 2. 해결책: 붕어빵 틀(Class) 만들기

변수를 하나하나 만드는 대신, **"학생 데이터를 담을 수 있는 공통된 틀"**을 하나 만듭니다.
이것이 바로 **클래스(Class)**입니다.

* **클래스(Class):** 붕어빵 기계 (설계도)
* **객체(Object/Instance):** 실제 붕어빵 (실체)

가장 기본적인 껍데기만 만들어 보겠습니다.

```python
# [문법] class 이름:
# 클래스 이름은 보통 첫 글자를 대문자로 씁니다. (Student)

class Student:
    pass 
    
# 'pass'는 "내용은 나중에 채울 테니 에러를 내지 말고 일단 넘어가라"는 뜻입니다.
# 이제 파이썬은 'Student'라는 설계도가 존재한다는 것을 알게 되었습니다.
```

---

### 3. 틀을 이용해 찍어내기 (객체 생성)

이제 만들어둔 붕어빵 틀(`Student`)을 사용해서 실제 팥 붕어빵, 슈크림 붕어빵(`student1`, `student2`)을 찍어냅니다.
이 과정을 **"인스턴스화(Instantiation)"**라고 합니다.

```python
# [중요] 클래스 이름 뒤에 괄호()를 붙이면 '객체'가 하나 툭 튀어나옵니다.

# 철수를 위한 공간(객체) 만들기
student1 = Student() 

# 영희를 위한 공간(객체) 만들기
student2 = Student()

# 확인해볼까요?
print(student1)
print(student2)
```

**실행 결과 해석:**
`<__main__.Student object at 0x...>`  
뒤에 나오는 복잡한 숫자(0x...)는 **메모리 주소**입니다.
두 결과의 주소가 다르다는 것은, `student1`과 `student2`가 같은 틀에서 나왔지만 **서로 완전히 다른 독립적인 존재**라는 뜻입니다.

---

### 4. 만들어진 객체에 데이터 담아보기

이제 `student1`과 `student2`라는 빈 붕어빵이 생겼으니, 여기에 각각 팥과 슈크림을 넣듯이 데이터를 넣어보겠습니다.
객체 안에 있는 변수에 접근할 때는 **점(.)**을 사용합니다.

```python
# student1(철수)만의 데이터 저장
student1.name = "철수"  # name이라는 변수를 그 자리에서 만들어 넣음
student1.score = 90

# student2(영희)만의 데이터 저장
student2.name = "영희"
student2.score = 85

# [결과 확인] 서로 섞이지 않고 잘 저장되었는지 볼까요?
print(f"1번 학생: {student1.name}, 점수: {student1.score}")
print(f"2번 학생: {student2.name}, 점수: {student2.score}")
```

> **핵심:** `student1.name`을 바꾼다고 해서 `student2.name`이 바뀌지 않습니다. 
> 마치 내 휴대폰에 저장된 사진을 지워도 친구 휴대폰 사진은 그대로인 것과 같습니다.
> 이것이 바로 **"객체의 독립성"**입니다.

---

2. **객체(`Object`)**: 설계도로 찍어낸 **실체 (붕어빵)**
3. **사용법**: `변수 = 클래스명()` 으로 만들고, `변수.속성 = 값` 으로 데이터를 넣는다.
