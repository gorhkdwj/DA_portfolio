### 1. 일일이 입력하던 귀찮음 해결하기 (`__init__`)

1교시 마지막에 했던 방식의 단점을 보셨나요?
객체를 만들고 나서, `.name = ...`, `.score = ...` 이렇게 **매번 수동으로** 넣어줘야 했습니다. 깜빡하고 점수를 안 넣으면 에러가 나겠죠?

이걸 해결하기 위해 **"객체가 생성될 때(initialize) 자동으로 실행되는 함수"**인 `__init__`을 사용합니다.

```python
class Student:
    # __init__ : 태어날 때 무조건 실행되는 마법의 함수
    # self : '방금 태어난 나 자신'을 가리킴 (필수!)
    def __init__(self, name, score):
        self.name = name   # 내(self) 이름표에 입력받은 name을 적어라
        self.score = score # 내(self) 점수판에 입력받은 score를 적어라
        print(f"응애! {self.name} 학생이 태어났습니다.")

# --- 비교 ---

# [옛날 방식] (X) - 귀찮음
# s1 = Student()
# s1.name = "철수" ...

# [새로운 방식] (O) - 생성과 동시에 데이터 입력!
# 괄호 안에 값을 던져주면, 알아서 __init__의 name, score 자리로 쏙 들어갑니다.
s1 = Student("철수", 90) 
s2 = Student("영희", 85)
```

> **핵심:** 이제 `Student("이름", 점수)` 한 줄이면, 데이터가 채워진 상태로 객체가 짠! 하고 나타납니다.

### 2. 행동 부여하기 (메서드 Method)

데이터(이름, 점수)만 있으면 엑셀과 다를 게 없죠. 이제 이 학생이 **'인사'**도 하고 **'공부'**도 하게 만들어봅시다.
클래스 안에 만드는 함수를 **메서드(Method)**라고 부릅니다.

⚠️ **주의할 점:** 메서드를 만들 때는 괄호 안에 **무조건 첫 번째로 `self`**를 써야 합니다. 그래야 "누가" 행동하는지 알 수 있거든요.

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    # [행동 1] 자기소개하기
    def introduce(self):
        # 그냥 name이라고 쓰면 안 됩니다! '내 이름'이니까 self.name
        print(f"안녕하세요, 저는 {self.name}이고, 점수는 {self.score}점입니다.")

    # [행동 2] 공부하기 (점수 올리기)
    def study(self, hours):
        # 공부한 시간만큼 점수 더하기
        self.score += hours 
        print(f"으랏차차! {hours}시간 공부해서 점수가 올랐어요.")
```

```python
# --- 사용해보기 ---

cheolsu = Student("철수", 90)

# 철수야, 소개해봐!
cheolsu.introduce() 
# 출력: 안녕하세요, 저는 철수이고, 점수는 90점입니다.

# 철수야, 5시간 공부하자!
cheolsu.study(5)    
# 출력: 으랏차차! 5시간 공부해서 점수가 올랐어요.

# 정말 올랐는지 확인
cheolsu.introduce() 
# 출력: 안녕하세요, 저는 철수이고, 점수는 95점입니다. (90 + 5)
```

### 🎓 2교시 핵심 요약

1. **`__init__`**: 객체가 태어날 때 **기본 스펙(속성)**을 세팅하는 곳입니다.
2. **`self`**: **"나 자신"**을 가리킵니다. 클래스 내부에서 변수나 함수를 쓸 때는 항상 `self.`을 붙여야 "내 것"을 씁니다.
3. **메서드**: 클래스 안에 만든 함수로, 객체의 **행동**을 정의합니다.
