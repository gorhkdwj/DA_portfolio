---

### 1. 문제 설명

> 각에서 0도 초과 90도 미만은 예각, 90도는 직각, 90도 초과 180도 미만은 둔각 180도는 평각으로 분류합니다. 각 angle이 매개변수로 주어질 때 예각일 때 1, 직각일 때 2, 둔각일 때 3, 평각일 때 4를 return하도록 solution 함수를 완성해주세요.
예각 : 0 < angle < 90
직각 : angle = 90
둔각 : 90 < angle < 180
평각 : angle = 180

### 2. 제한사항

* 0 < angle ≤ 180
* angle은 정수입니다.

### 3. 입출력 예

입출력 예 #1

angle이 70이므로 예각입니다. 따라서 1을 return합니다.
입출력 예 #2

angle이 91이므로 둔각입니다. 따라서 3을 return합니다.
입출력 예 #2

angle이 180이므로 평각입니다. 따라서 4를 return합니다.

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: (이 문제를 풀기 위해 필요한 핵심 아이디어를 한 줄로 적어두세요. 예: 리스트 컴프리헨션 사용, 스택 구조 활용 등)
---


```python
# [풀이 전략]
# 1. 정수를 입력받는다.
# 2. if문을 사용하여 구간별로 구분한다.
# 3. 구분한 구간별로 answer에 값을 할당한다.
# 4. 결과를 반환한다.

def solution(angle):
    if angle == 180 :
        answer = 4
    elif angle > 90 :
        answer = 3
    elif angle == 90 :
        answer = 2
    else :
        answer = 1
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(90))   # 예상: 2
print(solution(180)) # 예상: 4
print(solution(30)) # 예상: 1
```

# [Lv.1] 문제 제목 (짝수의 합)

- **출처**: 프로그래머스
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/120831)
- **카테고리**: for 반복문, if 조건문

---

### 1. 문제 설명

> 정수 n이 주어질 때, n이하의 짝수를 모두 더한 값을 return 하도록 solution 함수를 작성해주세요.

### 2. 제한사항

* 0 < n ≤ 1000
### 3. 입출력 예
입출력 예 #1

n이 10이므로 2 + 4 + 6 + 8 + 10 = 30을 return 합니다.
입출력 예 #2

n이 4이므로 2 + 4 = 6을 return 합니다.
---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: 
1. for 반복문과 range() 함수를 사용하여 n 이하의 모든 짝수를 더합니다.
2. if 조건문을 사용하여 짝수를 확인합니다.
---


```python
# [풀이 전략]
# 1. 정수 n을 입력받는다.
# 2. for 반복문과 range() 함수를 사용하여 1부터 n까지의 수를 순회한다.
# 3. if 조건문을 사용하여 i가 짝수인지 확인한다.
# 4. 짝수인 경우 answer에 i를 더한다.
# 5. 결과를 반환한다.

def solution(n):
    answer = 0
    for i in range (1,n+1):
        if i%2==0:
            answer+=i
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(8))   # 예상: 20
print(solution(100)) # 예상: 2550
```

# [Lv.1] 문제 제목 (배열의 평균값)

- **출처**: 프로그래머스 / 백준
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/120817)
- **카테고리**: 내장함수 / 산술연산 / 모듈

---

### 1. 문제 설명

> 정수 배열 numbers가 매개변수로 주어집니다. numbers의 원소의 평균값을 return하도록 solution 함수를 완성해주세요.

### 2. 제한사항

* 0 ≤ numbers의 원소 ≤ 1,000
* 1 ≤ numbers의 길이 ≤ 100
* 정답의 소수 부분이 .0 또는 .5인 경우만 입력으로 주어집니다.

### 3. 입출력 예

입출력 예 #1

numbers의 원소들의 평균 값은 5.5입니다.
입출력 예 #2

numbers의 원소들의 평균 값은 94.0입니다.

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: 
* 평균을 구하기 위한 sum() 함수와 len() 함수를 사용한다.
---


```python
# [풀이 전략]
# 1. numbers를 입력받는다.
# 2. numbers의 합을 구한다.
# 3. numbers의 개수를 구한다.
# 4. numbers의 합을 numbers의 개수로 나눈다.
# 5. 결과를 반환한다.

def solution(numbers):
    num_sum = sum(numbers)
    num_len = len(numbers)
    answer = num_sum/num_len
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution([1,2,3,4,5,6]))   # 예상: 3.5
print(solution([5,5])) # 예상: 5

# 아래와 같이 statistics 모듈을 사용하여 풀 수도 있었다.
from statistics import mean 

def solution(numbers):
    answer = mean(numbers)
    return answer
```

> 정수 num이 짝수일 경우 "Even"을 반환하고 홀수인 경우 "Odd"를 반환하는 함수, solution을 완성해주세요.

### 4. 문제 풀이 핵심 (선택 사항)

```python
# [풀이 전략]
# 1. num을 입력받는다.
# 2. if문을 사용하여 나머지 연산을 통해 짝수와 홀수를 판별한다.
# 3. 알맞는 결과를 반환한다.

def solution(num):
    if num%2==0:
        answer = 'Even'
    else:
        answer = 'Odd'
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(2, 3))   # 예상: -1
print(solution(100, 2)) # 예상: 98
```

### 1. 문제 설명

> 정수를 담고 있는 배열 arr의 평균값을 return하는 함수, solution을 완성해보세요.

### 2. 제한사항

* arr은 길이 1 이상, 100 이하인 배열입니다.

# [Lv.1] 문제 제목 (자릿수 더하기)

- **출처**: 프로그래머스
- **링크**: [문제 바로가기](https://school.programmers.co.kr/learn/courses/30/lessons/12931)
- **카테고리**: 리스트 / 문자열 / 해시 (관련 개념 태그)

---

### 1. 문제 설명

> (여기에 문제의 지문 내용을 복사해서 붙여넣으세요. 인용구(>)를 쓰면 깔끔합니다.)
>
> 정수 num1과 num2가 주어질 때, num1에서 num2를 뺀 값을 return하도록 solution 함수를 완성해주세요.

### 2. 제한사항

* (여기에 제한사항을 적으세요)
* -50000 ≤ num1 ≤ 50000

### 3. 입출력 예

| num1 | num2 | result |
| :--- | :--- | :--- |
| 2 | 3 | -1 |
| 100 | 2 | 98 |

---

### 4. 문제 풀이 핵심 (선택 사항)

* **Key Point**: (이 문제를 풀기 위해 필요한 핵심 아이디어를 한 줄로 적어두세요. 예: 리스트 컴프리헨션 사용, 스택 구조 활용 등)
---


```python
# [풀이 전략]
# 1. n을 입력받는다.
# 2. 빈 리스트를 만들고 for 문과 range(len())구문을 통해 반복 범위 지정.
# 3. 각 자리수만큼 10을 i제곱하여 나눈 나머지-그 아랫자리를 10의 (i-1)제곱으로 나눈 나머지
# 4. 3.의 값을 10의 (i-1)제곱으로 나눠주면 해당 자리의 숫자만 추출 가능하다.
# 5. 각 자리의 숫자를 리스트에 append하여 sum 하면 각 자릿수의 합을 구할 수 있다.

def solution(n):
    y=str(n)
    lst=[]
    for i in range(1,len(y)+1):
        x=((n%(10**i))-(n%(10**(i-1))))/(10**(i-1))
        lst.append(x)
    answer=sum(lst)
    return answer


# 12345
# 10의 1제곱으로 나눈 나머지 - 10의 0제곱으로 나눈 나머지
# 10의 2제곱(100)으로 나눈 나머지 - 10의 1제곱으로 나눈 나머지

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(22351))   # 예상: 13
print(solution(12345)) # 예상: 15
```
