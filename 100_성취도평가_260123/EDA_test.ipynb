```python
import pandas as pd
orders = pd.read_csv("bootcamp_orders.csv")
customers = pd.read_csv("bootcamp_customers.csv")

# Q19 (Q18 단계 복사)
orders_clean = orders.copy()

orders_clean = orders_clean.drop_duplicates(subset=["order_id"], keep="first")

orders_clean["discount_rate"] = orders_clean["discount_rate"].fillna(0)

orders_clean["revenue"] = orders_clean["unit_price"] * orders_clean["quantity"] * (1-orders_clean["discount_rate"])

# Q19-1
merged = orders_clean.merge(customers, on='customer_id', how='left')
merged.head()

# Q19-2
merged_clean = merged.copy()
merged_clean["city"] = merged_clean["city"].fillna("UNKNOWN")

# Q19-3
city_agg = merged_clean.groupby("city").agg(
    order_cnt = ("order_id", "nunique"),
    customer_cnt = ("customer_id", "nunique"),
    revenue_sum = ("revenue", "sum")
)
city_agg_sorted = city_agg.sort_values(
    by = ["revenue_sum", "city"],
    ascending=[False, True]
)

print(city_agg_sorted)
```
