### 실습 문제 1 : 리스트의 합 `list_sum`

**문제 설명**

- 정수로만 이루어진 리스트 `nums`를 매개변수로 받아 **모든 요소의 합**을 반환하는 함수를 만드세요.
- 리스트가 비어 있다면 `0`을 반환해야 합니다.

**예시 실행**

```python
print(list_sum([1, 2, 3, 4]))  # 10
print(list_sum([]))            # 0
```

````python
def list_sum(nums) :
  if len(nums)==0:
    return 0
  return sum(nums)
print(list_sum([1,2,3,4]))
print(list_sum([]))
````

### 실습 문제 2 : 단어 빈도 세기 함수 `word_count`

**문제 설명**

- 문자열을 매개변수로 받아, 공백을 기준으로 단어를 분리합니다.
- 각 단어가 몇 번 나왔는지 **딕셔너리** 형태로 반환하세요.
- 대소문자는 구분하지 않으며, 모두 소문자로 변환한 뒤 계산하세요.

**예시 실행**

```python
print(word_count("Apple banana apple Orange orange banana apple"))
# {'apple': 3, 'banana': 2, 'orange': 2}
```

````python
def word_count(a):
  a=a.lower()
  words=a.split()

  word_dict={}
  for word in words:
    if not word in word_dict:
      word_dict[word]=1
    else:
      word_dict[word] += 1
  return word_dict

print(word_count("Apple banana apple Orange orange banana apple"))
````

### 실습 문제 3 : 학점 계산 함수 `grade_report`

**문제 설명**

- 학생 이름과 점수가 들어있는 딕셔너리를 매개변수로 받으세요.
- 점수를 기준으로 **학점(A, B, C, D, F)**을 매겨서 새 딕셔너리로 반환하세요.
- 학점 기준은 다음과 같습니다.

| 점수 범위 | 학점 |
| --- | --- |
| 90 이상 | A |
| 80~89 | B |
| 70~79 | C |
| 60~69 | D |
| 60 미만 | F |

**예시 실행**

```python
scores = {"Alice": 85, "Bob": 92, "Charlie": 67, "Dave": 58}
print(grade_report(scores))
# {'Alice': 'B', 'Bob': 'A', 'Charlie': 'D', 'Dave': 'F'}
```

````python
# 성적 세트가 들어올 때마다 각 학생의 학점을 출력해주는 함수 만들기!
def grade_report(s):
  new_dict={}
  for student in s:
    if s[student] >= 90:
      new_dict[student]="A"
    elif s[student] >= 80:
      new_dict[student]="B"
    elif s[student] >= 70:
      new_dict[student]="C"
    elif s[student] >= 60:
      new_dict[student]="D"
    else:
      new_dict[student]="F"
  return new_dict




scores = {"Alice": 85, "Bob": 92, "Charlie": 67, "Dave": 58}
print(grade_report(scores))
````

````python
# 함수 생성
# 각 점수 Values 딕셔너리에서 뽑아 학점과 비교해야 함.
# 매개변수: 이름과 점수가 있는 딕셔너리
# 점수를 기준으로 학점을 매겨 "새 딕셔너리로" 반환.
````

````python
# AI의 더 깔끔한 코드 피드백

def grade_report(scores):  # 변수명도 s보다 scores가 더 명확하겠죠?
    new_dict = {}

    # [핵심] 키(name)와 값(score)을 한 번에 꺼냅니다!
    for name, score in scores.items(): # item() 사용으로
        if score >= 90: # score 변수 활용하여 가독성이 더 좋아짐!
            new_dict[name] = "A"
        elif score >= 80:
            new_dict[name] = "B"
        elif score >= 70:
            new_dict[name] = "C"
        elif score >= 60:
            new_dict[name] = "D"
        else:
            new_dict[name] = "F"

    return new_dict
````
