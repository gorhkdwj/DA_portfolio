### 문제 1

학생의 이름과 나이를 저장하는 `Student` 클래스를 만들어 보세요.

**조건**

- `__init__` 메서드를 사용하세요.
- 이름(name)과 나이(age)를 속성으로 저장하세요.
- 학생의 정보를 출력하는 `print_info()` 메서드를 만드세요.

**출력 예시**

```
이름: 민수, 나이: 15
```

````python
class Student:

  def __init__(self, name, age):
    self.name = name
    self.age = age

  def info(self):
    print(f"이름: {self.name}, 나이: {self.age}")

s1 = Student("민수", 15)
s1.info()
````

### 문제 2

사각형을 표현하는 `Rectangle` 클래스를 만들어 보세요.

**조건**

- 가로(width), 세로(height)를 인스턴스 변수로 가집니다.
- 이 인스턴스 변수들은 밖에서 직접 접근이 불가능합니다. (힌트 : `__` )
- 사각형의 넓이를 반환하는 `get_area()` 메서드를 만드세요.
- 메서드를 호출하여 넓이를 출력하세요.

**힌트**

- 넓이 = 가로 × 세로

````python
class Rectangle:

  def __init__ (self, width, height):
    self.width = width
    self.height = height

  def get_area(self):
    area = self.width * self.height
    return area

s1 = Rectangle(3,4)
print(s1.get_area())
````

### 문제 5

아래 코드의 출력 결과를 예측하고, 왜 그렇게 되는지 설명하세요.

```python
class Box:
    def __init__(self, items):
        self.items = items

box1 = Box([1, 2, 3])
box2 = box1

box2.items.append(4)

print(box1.items)
```

**힌트**

- `box1`과 `box2`가 가리키는 객체를 생각해 보세요.
- 리스트는 변경 가능한 객체(mutable)입니다.

````python
class Box:
    def __init__(self, items):
        self.items = items

box1 = Box([1, 2, 3])
box2 = box1

box2.items.append(4)

print(box1.items)
print(box2.items)
````

### 문제 6

아래 코드를 실행했을 때 **출력 결과를 정확히 예측**하고,

왜 그런 결과가 나오는지 **클래스 변수 / 인스턴스 변수 관점에서 설명**하세요.

```python
class Score:
    total = 0

    def __init__(self, value):
        self.value = value
        Score.total += value

    def add(self, x):
        self.value += x
        Score.total += x

a = Score(10)
b = Score(20)

a.add(5)
b.add(3)

print(a.value)
print(b.value)
print(Score.total)
```

**힌트**

- `self.value`는 어디에 저장되는 값인가요?
- `Score.total`은 언제, 어디에서 증가하나요?

````python
class Score:
    total = 0

    def __init__(self, value):
        self.value = value
        Score.total += value

    def add(self, x):
        self.value += x
        Score.total += x

a = Score(10)
b = Score(20)

a.add(5)
b.add(3)

print(a.value)
print(b.value)
print(Score.total)
````

### 문제 7

아래 코드에는 **의도와 다르게 동작하는 부분**이 있습니다.

**의도**

- 각 학생 객체가 **자기만의 점수 리스트**를 가져야 한다.

코드를 실행했을 때 실제 출력 결과를 예측하고,

왜 이런 문제가 발생하는지 설명한 뒤, **올바르게 수정하세요.**

```python
class Student:
    scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)

```

**힌트**

- `scores`는 클래스 변수인가요, 인스턴스 변수인가요?
- 리스트는 어떤 특징을 가진 객체인가요?

````python
# 각각 새로운 리스트를 생성해서 append해야하는데 전부 원몬 리스트에 append중
# 새로운 리스트를 어떻게 만들어줄까...?
# init을 사용해 []를 클래스 변수가 아닌 인스턴스 변수로 만든다

class Student:
    def __init__(self):
      self.scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)
````
