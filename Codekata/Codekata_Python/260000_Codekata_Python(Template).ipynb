# [풀이 전략]
# 1. 두 수를 입력받는다.
# 2. 뺄셈 연산자(-)를 사용한다.
# 3. 결과를 반환한다.

def solution(num1, num2):
    answer = num1 - num2
    return answer

# 테스트 (직접 실행해서 결과를 확인했다는 증거)
print(solution(2, 3))   # 예상: -1
