```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def print_info(self):
        print(f"이름: {self.name}, 나이: {self.age}")

p1 = Student("민수", 15)
p1.print_info()
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def get_area(self):
        area = self.width * self.height
        return area

R1 = Rectangle(4,5)
print(f"{R1.get_area()}입니다.")
```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog = Dog("초코", 3)
print(dog.name)
print(dog.age)
```

```python
class Counter:
    count = 0
    def add(self):
        self.count += 1

a = Counter()
b = Counter()

a.add()
a.add()
b.add()

print(a.count)
print(b.count)
```

```python
class Box:
    def __init__(self, items):
        self.items = items

box1 = Box([1, 2, 3])
box2 = box1

box2.items.append(4)

print(box1.items)
```

```python
class Score:
    total = 0

    def __init__(self, value):
        self.value = value
        Score.total += value

    def add(self, x):
        self.value += x
        Score.total += x

a = Score(10)
b = Score(20)

a.add(5)
b.add(3)

print(a.value)
print(b.value)
print(Score.total)
```

```python
class Student:
    def __init__(self):
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)
```
