# 리스트 실습문제

```python
#실습문제 1
cities = ["Seoul", "Busan", "Incheon", "Gwangju", "Daejeon"]

# TODO 1) 리스트에서 가운데 3개 도시만 잘라 sub_cities 에 저장
# TODO 2) sub_cities 의 순서를 거꾸로 뒤집어 reversed_cities 에 저장
# TODO 3) reversed_cities 에 "Ulsan" 을 맨 뒤에 추가
# TODO 4) 최종 리스트를 출력

# 1-1.
sub_cities=cities[1:4]
print(sub_cities)

# 1-2.
cities = ["Seoul", "Busan", "Incheon", "Gwangju", "Daejeon"]
sub_cities=cities.pop(0)and cities.pop(-1)
sub_cities=cities
print(sub_cities)

# 2-1. 복사본을 만들어서, reverse 시키기
reversed_cities=[]
reversed_cities=sub_cities.copy()
reversed_cities.reverse()
print(reversed_cities)

# 2-2. sub_cities의 순서를 거꾸로 뒤집어 reversed_cities
# for 문으로 거꾸로 돌면서 붙이기(append)
reversed_cities=[]
for i in range(len(sub_cities)-1,-1,-1):
  reversed_cities.append(sub_cities[i])
print(reversed_cities)

# 3.
reversed_cities.append("Ulsan")
print(reversed_cities)
```

```python
# 실습문제 2번 - List Comprehension XX
result=[]
for i in range(1,31):
  if i%3==0 and i%2!=0:
    result.append(i**2)
print(result)
```

```python
# 실습문제 2번 - List Comprehension
result=[]
result=[num**2 for num in range(1,31) if (num%3==0 and num%2!=0)]
print(result)
```

```python
# 실습문제 3번
num_list=(input("정수를 입력해주세요 (공백 구분): ").split())
new_list=[]
for _ in num_list:
  new_list.append(int(_))
print(f"최댓값: {max(new_list)}")
print(f"최솟값: {min(new_list)}")
print(f"평균: {sum(new_list)/len(new_list)}")
```

# 딕셔너리 실습문제

```python
# 실습문제 1번

phonebook={}

for _ in range(3):
  name=input("이름을 입력하세요: ")
  num=input("전화번호를 입력하세요: ")
  phonebook[name]=num

for key in phonebook:
  print(f"{key} : {phonebook[key]}")

name_del=input("삭제할 이름: ")
if name_del in phonebook:
  phonebook.pop(name_del)
  print(f"{name_del} 연락처가 삭제되었습니다.")
  print(f"최종 연락처: {phonebook}")
else:
  ("해당 이름이 없습니다.")
```
