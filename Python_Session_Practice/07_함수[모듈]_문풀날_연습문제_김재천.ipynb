### 문제 1

두 개의 숫자를 받아서 **더 큰 값을 반환하는 함수** `bigger(a, b)`를 만들어 보세요.

예를 들어 `bigger(3, 5)`를 실행하면 `5`가 반환되어야 합니다.

````python
def bigger(num1, num2):
  if num1>num2:
    return num1
  elif num1<num2:
    return num2
  else:
    print("두 수가 같습니다!")
    return

# map 함수에 대해 재학습! map object 개념에 대해서도 알 수 있었다.
a,b=map(int,input("두 수를 입력해주세요(공백으로 구분).: ").split())

print(bigger(a,b))
````

### 문제 2

### 문제 3

예를 들어 `square(4)`는 `16`을 반환해야 합니다.

### 문제 4

### 문제 5

아래 함수는 리스트의 첫 번째 값을 0으로 바꾸려고 합니다.

하지만 실행 결과를 보면 원래 리스트가 바뀌지 않습니다.

그 이유를 설명하고, 리스트가 실제로 바뀌도록 코드를 수정해 보세요.

```python
def change_first(x):
    x = [0, 2, 3]

nums = [1, 2, 3]
change_first(nums)
print(nums)

```

````python
def change_first(x):
    x = [0, 2, 3]
    return x

nums = [1, 2, 3]
nums=change_first(nums)
print(nums)

# 위 코드의 의도대로 nums의 값을 change_first(nums)값으로 바꾸고 싶다면
# nums = change_first(nums) 구문을 통해 nums자체 값을 함수의 반환값으로 바꿔줘야 함.
# change_first(nums)는 함수 내부에서 값을 처리하고 끝내기 때문에 scope가 달라 원 변수의 값도 다시 바꿔줘야 함.
# 따라서 함수 내부에 return x를 삽입하고, nums=change_first(nums)를 추가한다.
````

````python
def change_first(x):
  x[0] = 0
  return x

nums = [1, 2, 3]
change_first(nums)
print(nums)
# 위 코드는 인덱싱을 통해 원래 nums가 가지고 있던 list에 직접 접근하여 해당 값을 교체함 같은 scope
# 함수에서 따로 list를 생성하여 바꾼 것이 아닌, 원본이 바뀌었음.
# 따라서 추가 구문 필요 없이 정상 작동
# (만일 원본 list를 변경하면 안된다면 주의!)
````

하지만 실행 결과는 예상과 다릅니다.

**출력 결과를 먼저 예측**하고, 왜 그런 결과가 나오는지 설명하세요.

```python

    lst = [0, 0, 0]

print(nums)

```

### 문제 9

`math` 모듈을 사용하여 **원의 넓이**를 계산하는 프로그램을 만들어 보세요.

사용 방법:

- 반지름을 입력받는다.
- `math.pi`를 사용하여 원의 넓이(π × r × r)를 계산하는 함수를 만든다.
- 계산된 넓이를 출력한다.

예시 상황:

반지름이 3이라면, 원의 넓이를 계산해서 출력해야 합니다.

````python
from math import pi
# pi값(3.141592...)을 가져오는 모듈

def area(r):
  return r*(pi**2)

radius=int(input("반지름을 입력하세요"))
area=area(radius)
print(f"넓이: {area}")
````
