```python
## 막힐 때 사용할 답안지! (가급적 보지 말고 풀어보자)


# ==========================================
# Project: PJ04_Secure_OOP_Portfolio
# 목표: 캡슐화(Private)와 상속(Inheritance)을 적용한 안전한 자산 관리 시스템 구현
# 작성자: (본인의 이름을 적으세요)
# ==========================================

class Portfolio:
    """
    [Mission 1] 클래스 변수 설정
    모든 계좌에 공통으로 적용되는 기본 거래 수수료율(commission_rate)을 1%(0.01)로 설정하세요.
    """
    commission_rate = 0.01

    def __init__(self, owner, initial_cash):
        self.owner = owner
        self.holdings = {}  # 보유 주식: {'종목명': {'qty': 0, 'avg_price': 0.0}}
        
        """
        [Mission 2] 캡슐화 (Encapsulation)
        초기 자본금(initial_cash)을 외부에서 함부로 접근하지 못하도록
        Private 변수(__balance)로 만들어 할당하세요.
        """
        self.__balance = initial_cash
        
        print(f"[{self.owner}] 포트폴리오 생성 완료 (잔고: {self.__balance:,.0f}원)")

    def get_balance(self):
        """
        [Mission 3] Getter 메서드
        외부에서 __balance 값을 읽을 수(Read-only) 있도록 반환하는 메서드를 작성하세요.
        """
        return self.__balance

    def deposit(self, amount):
        """계좌 입금 메서드"""
        if amount > 0:
            self.__balance += amount
            print(f"[{self.owner}] {amount:,.0f}원 입금 (현재 잔고: {self.__balance:,.0f}원)")
        else:
            print("입금액은 0보다 커야 합니다.")

    def buy(self, ticker, qty, price):
        """
        [Mission 4] 매수 로직 구현
        1. 매수 금액(amount)과 수수료(fee)를 계산하여 총 비용(total_cost) 산출
        2. 잔고(__balance)가 충분한지 확인
        3. 보유 목록(holdings) 갱신 및 이동평균법 평단가 계산
        """
        # 1. 비용 계산
        amount = qty * price
        fee = amount * self.commission_rate  # 클래스 변수 활용
        total_cost = amount + fee

        # 2. 잔고 확인 (Private 변수 접근)
        if total_cost > self.__balance:
            print(f"[매수 실패] 잔고 부족 (필요: {total_cost:,.0f}원 / 보유: {self.__balance:,.0f}원)")
            return

        # 3. 로직 구현 (TODO: 아래 pass를 지우고 작성하세요)
        # 3-1. 잔고 차감
        self.__balance -= total_cost
        
        # 3-2. holdings에 종목이 없으면 초기화
        if ticker not in self.holdings:
            self.holdings[ticker] = {'qty': 0, 'avg_price': 0.0}

        # 3-3. 평단가 계산 (이동평균법)
        # new_avg = ((기존평단 * 기존수량) + (매수단가 * 매수수량)) / 총수량
        old_qty = self.holdings[ticker]['qty']
        old_avg = self.holdings[ticker]['avg_price']
        
        new_avg = ((old_avg * old_qty) + (price * qty)) / (old_qty + qty)

        # 3-4. 수량 증가 및 평단가 갱신
        self.holdings[ticker]['qty'] += qty
        self.holdings[ticker]['avg_price'] = new_avg
        
        print(f"[매수 체결] {ticker} {qty}주 @ {price:,.0f}원 (수수료: {fee:,.0f}원)")

    def sell(self, ticker, qty, price):
        """
        [Mission 5] 매도 로직 구현
        1. 보유 수량 확인
        2. 수수료를 제외한 정산 금액(net_income) 계산
        3. 실현 손익(profit) 계산 및 반환
        """
        # 1. 예외 처리
        if ticker not in self.holdings or self.holdings[ticker]['qty'] < qty:
            print(f"[매도 실패] {ticker} 수량이 부족합니다.")
            return 0

        # 2. 정산 금액 계산
        revenue = qty * price
        fee = revenue * self.commission_rate
        net_income = revenue - fee

        # 3. 로직 구현 (TODO: 아래 pass를 지우고 작성하세요)
        # 3-1. 실현 손익 계산: 정산금액 - (평단가 * 수량)
        profit = net_income - (self.holdings[ticker]['avg_price'] * qty)
        
        # 3-2. 잔고 증가 및 수량 감소
        self.__balance += net_income
        self.holdings[ticker]['qty'] -= qty

        print(f"[매도 체결] {ticker} {qty}주 처분 (수익: {profit:,.0f}원 / 수수료: {fee:,.0f}원)")
        return profit


class VIPPortfolio(Portfolio):
    """
    [Mission 6] 상속과 오버라이딩 (Inheritance & Overriding)
    Portfolio 클래스를 상속받아 VIP 전용 클래스를 만드세요.
    VIP는 거래 수수료(commission_rate)가 0.0 (무료)입니다.
    """
    # TODO: 부모의 commission_rate를 0.0으로 덮어쓰세요.
    commission_rate = 0.0

    def introduce(self):
        print(f"💎 {self.owner} VIP님 환영합니다! 수수료 면제 혜택이 적용됩니다.")


# ==========================================
# [Main] 테스트 실행 코드
# ==========================================
if __name__ == "__main__":
    print("=== 1. 일반 계좌 테스트 (수수료 1%) ===")
    p1 = Portfolio("개미", 10000000)
    p1.buy("Samsung", 10, 80000)  # 수수료 8,000원 발생 예상
    
    print("\n=== 2. VIP 계좌 테스트 (수수료 0%) ===")
    p2 = VIPPortfolio("워렌버핏", 500000000)
    p2.introduce()
    p2.buy("Samsung", 10, 80000)  # 수수료 0원 예상
    
    print("\n=== 3. 보안성(캡슐화) 테스트 ===")
    print(f"개미 잔고 확인(Getter): {p1.get_balance():,.0f}원")
    
    # 아래 주석을 풀고 실행했을 때 에러가 나야 성공입니다!
    # print(p1.__balance) 
    
    # 아래 주석을 풀고 실행했을 때 잔고가 바뀌지 않거나 에러가 나야 성공입니다!
    # p1.__balance = 10000000000
    # print(f"해킹 시도 후 잔고: {p1.get_balance():,.0f}원")
```
