### 🚀 단계별 구현 가이드

#### Step 1. [설계] 보안이 강화된 클래스 뼈대 구축 (Encapsulation)

* `Portfolio` 클래스 정의.
* **핵심:** `self.balance` 대신 `self.__balance` (Private)를 사용하여 외부 접근 차단.
* 잔고 확인을 위한 `get_balance()` (Getter) 메서드와 안전한 `deposit()` 메서드 구현.

#### Step 2. [구현] 수수료 정책이 포함된 매매 로직 (Class Variable)

* 클래스 변수 `commission_rate = 0.01` (1%) 선언.
* `buy` / `sell` 메서드 구현 시, 이 수수료율을 반영하여 총 비용/수익 계산.
* 어제의 '이동평균법 평단가 계산' 로직을 메서드 내부로 이식.

#### Step 3. [확장] 상속을 이용한 VIP 계좌 개설 (Inheritance)

* `Portfolio`를 상속받는 `VIPPortfolio` 클래스 정의.
* **Overriding:** 부모의 기능은 그대로 두되, `commission_rate`만 `0.0`으로 재정의하거나, `buy` 메서드를 수정하여 혜택 적용.
* 코드 중복 없이 기능이 확장되는 '상속의 마법' 체험.

#### Step 4. [검증] 시뮬레이션 및 다형성 테스트

* `일반_개미(Portfolio)`와 `워렌_버핏(VIPPortfolio)` 두 개의 인스턴스 생성.
* 동일한 금액으로 주식을 매수했을 때, 수수료 차이로 인해 최종 잔고가 달라지는지 확인.
* 외부에서 `개미.__balance = 0`으로 해킹을 시도했을 때 에러가 발생하는지 확인.

```python
# ==========================================
# Project: PJ04_Secure_OOP_Portfolio
# 작성자: (본인의 이름을 적으세요)
# ==========================================

class Portfolio:
    """
    [Mission 1] 클래스 변수 설정
    - 모든 계좌에 공통으로 적용되는 '기본 수수료율(commission_rate)'을 1%(0.01)로 설정하세요.
    """
    # TODO: 여기에 클래스 변수를 선언하세요.
    commission_rate = 0.01
    

    def __init__(self, owner, initial_cash):
        self.owner = owner
        self.holdings = {}  # 보유 주식: {'종목명': {'qty': 0, 'avg_price': 0.0}}
        
        """
        [Mission 2] 캡슐화 (Encapsulation) - 생성자
        - 초기 자본금(initial_cash)을 외부에서 함부로 접근하지 못하도록
        Private 변수(__balance)로 만들어 할당하세요.
        """
        # TODO: self.__balance 를 초기화하세요.
        self.__balance = initial_cash
        
        
        print(f"[{self.owner}] 포트폴리오 생성 완료")

    def get_balance(self):
        """
        [Mission 3] Getter 메서드
        - 캡슐화된 잔고(__balance)를 외부에서 읽을 수만 있게(Read-only) 반환하세요.
        """
        # TODO: 잔고를 return 하세요.
        return self.__balance

    def deposit(self, amount):
        """입금 메서드"""
        if amount > 0:
            # TODO: 입력받은 amount만큼 내부 잔고(__balance)를 증가시키세요.
            self.__balance += amount
            
            print(f"[{self.owner}] {amount:,.0f}원 입금 완료")
        else:
            print("입금액은 0보다 커야 합니다.")

    def buy(self, ticker, qty, price):
        """
        [Mission 4] 매수 로직 구현 (가장 중요!)
        
        [구현 가이드]
        1. 매수 금액(amount) = 수량 * 가격
        2. 수수료(fee) = 매수 금액 * 클래스 변수(commission_rate)
        3. 총 비용(total_cost) = 매수 금액 + 수수료
        
        4. [검증] 잔고(__balance)가 총 비용보다 적으면 "잔고 부족" 출력 후 리턴
        
        5. [상태 변경]
        - 잔고에서 총 비용 차감
        - self.holdings에 해당 종목이 없으면 빈 딕셔너리 {'qty':0, 'avg_price':0.0} 초기화
        
        6. [평단가 계산 - 이동평균법]
        - 이전 수량, 이전 평단가 가져오기
        - 새 평단가 = ((이전평단 * 이전수량) + (매수단가 * 매수수량)) / (이전수량 + 매수수량)
        
        7. [최종 저장]
        - 해당 종목의 수량 증가 (+qty)
        - 해당 종목의 평단가 갱신 (=새 평단가)
        """
        # TODO: 위 가이드에 따라 매수 로직을 처음부터 끝까지 작성하세요.
        ## 매수금액, 수수료, 총 비용 정의
        ## 매수금액 = 수량 * 가격
        ## 수수료 = 매수금액 * 수수료율
        ## 총 금액 = 매수금액 + 수수료
        
        amount = qty * price
        fee = amount * self.commission_rate
        total_cost = amount + fee

        ## 잔고와 총 비용 비교
        ## 적으면 잔고 부족

        if self.__balance < total_cost:
            print(f"[매수 실패] 잔고 부족.")
            print(f"필요 금액: {total_cost}원.")
            print(f"현재 잔고: {self.__balance}원.")
            print(f"필요 금액: {total_cost - self.__balance}원.")
            return

        ## 잔고가 총 비용보다 많을 시 잔고에서 총 비용을 빼고
        ## 해당 티커가 보유 주식에 있는 지 확인하고, 없으면 추가, 있으면 추가 후 평단가 수정
        else:
            self.__balance -= total_cost

        ## holdings에 종목이 없으면 초기화(있으면 그대로 진행)
        if ticker not in self.holdings:
            self.holdings[ticker] = {'qty': 0, 'avg_price':0.0}

        ## 평단가 계산
        ## new_avg = ((기존평단 + 기존수량) + (매수단가 * 매수수량) / (기존수량 + 매수수량))
        old_qty = self.holdings[ticker]['qty']
        old_avg = self.holdings[ticker]['avg_price']

        new_avg = ((old_avg * old_qty) + (price * qty)) / (old_qty + qty)

        ## 수량 증가 및 평단가 갱신
        self.holdings[ticker]['qty'] += qty
        self.holdings[ticker]['avg_price'] = new_avg

        print(f"[매수 체결] {ticker} {qty}주")
        print(f"주 당 {price:,.0f}원 (수수료: {fee:,.0f}원)")
        
    def sell(self, ticker, qty, price):
        """
        [Mission 5] 매도 로직 구현
        
        [구현 가이드]
        1. [검증] 보유 수량이 없거나 매도하려는 수량보다 적으면 "실패" 출력 후 0 반환
        
        2. [계산]
        - 총 판매금(revenue) = 수량 * 가격
        - 수수료(fee) = 총 판매금 * 클래스 변수(commission_rate)
        - 정산 금액(net_income) = 총 판매금 - 수수료
        
           - 실현 손익(profit) = 정산 금액 - (내 평단가 * 매도 수량)
        
        3. [상태 변경]
        - 잔고(__balance)에 정산 금액 더하기
        - 보유 수량(holdings)에서 매도 수량만큼 빼기
        
        4. [반환] profit 반환
        """
        # TODO: 위 가이드에 따라 매도 로직을 작성하세요.

        ## 보유수량이 적으면 걸러준다.
        if self.holdings[ticker]['qty'] < qty:
            print(f"[매도 실패] {ticker} 수량이 부족합니다.")
            print(f"매도 시도: {qty}주")
            print(f"보유 수량: {self.holdings[ticker]['qty']}주")
            return 0

        ## 총 판매금, 수수료, 정산 금액
        revenue = qty * price
        fee = revenue * self.commission_rate
        net_income = revenue - fee

        ## 실현 손익 계산: 정산금액 - (평단가 * 수량)
        profit = net_income - (self.holdings[ticker]['avg_price']*qty)

        ## 잔고 증가 및 수량 감소
        self.__balance += net_income
        self.holdings[ticker]['qty'] -= qty     
        
        
        print(f"[매도 체결] {ticker} {qty}주 처분 (수익: {profit:,.0f}원 / 수수료: {fee:,.0f}원)") # 수익 변수명을 포맷팅에 넣으세요
        return profit # 계산된 profit을 반환하도록 수정하세요.


# ==========================================
# [Mission 6] 상속 (Inheritance)
# - Portfolio 클래스를 상속받는 'VIPPortfolio' 클래스를 정의하세요.
# - VIP는 수수료(commission_rate)가 0.0 (무료)입니다.
# - (선택) introduce 메서드를 추가하여 VIP만의 인사를 출력해보세요.
# ==========================================

# TODO: class VIPPortfolio ... 작성
class VIPPortfolio(Portfolio):
    # 수수료 0
    commission_rate = 0.0

    def introduce(self):
        print(f"{self.owner} VIP님 환영합니다! 수수료 면제 혜택이 적용됩니다.")

# ==========================================
# [Main] 통합 테스트 실행 코드
# ==========================================
if __name__ == "__main__":
    print("\n=== 1. 일반 계좌 매수 테스트 (수수료 1%) ===")
    p1 = Portfolio("개미", 10000000)
    # 초기 잔고 확인
    print(f"초기 잔고: {p1.get_balance():,.0f}원")
    
    # 매수: 10주 * 80,000원 = 800,000원
    # 수수료: 8,000원 (1%)
    # 총 비용: 808,000원 차감 예상
    p1.buy("Samsung", 10, 80000) 
    print(f"매수 후 잔고: {p1.get_balance():,.0f}원") # 예상: 9,192,000원
    

    print("\n=== 2. 매도(Sell) 테스트 (수익 실현) ===")
    # 5주 * 90,000원 = 450,000원
    # 수수료: 4,500원
    # 입금액: 445,500원
    # 원금(평단 8만): 400,000원
    # 예상 수익: 45,500원
    profit = p1.sell("Samsung", 5, 90000)
    
    print(f"실현 수익: {profit:,.0f}원")
    print(f"매도 후 잔고: {p1.get_balance():,.0f}원") # 예상: 9,192,000 + 445,500 = 9,637,500원
    print(f"남은 수량: {p1.holdings['Samsung']['qty']}주")


    print("\n=== 3. VIP 계좌 테스트 (수수료 0%) ===")
    try:
        p2 = VIPPortfolio("워렌버핏", 500000000)
        p2.introduce() # 인사말 확인
        
        # VIP 매수: 수수료 없이 딱 80만원만 나가야 함
        p2.buy("Samsung", 10, 80000)
        print(f"VIP 매수 후 잔고: {p2.get_balance():,.0f}원") # 예상: 499,200,000원
        
        # VIP 매도: 수수료 없이 전액 입금되어야 함
        # 5주 * 90,000 = 450,000원 (수수료 0)
        # 원금 400,000원
        # 예상 수익: 50,000원
        vip_profit = p2.sell("Samsung", 5, 90000)
        print(f"VIP 실현 수익: {vip_profit:,.0f}원")
        
    except NameError:
        print("!! VIPPortfolio 클래스가 아직 구현되지 않았습니다 !!")


    print("\n=== 4. 보안성(캡슐화) 테스트 ===")
    try:
        print(p1.__balance) # 에러 발생해야 함
    except AttributeError:
        print("✅ 성공! 외부에서 __balance에 접근할 수 없습니다.")
    except Exception as e:
        print(f"❌ 실패! 예상치 못한 에러: {e}")
```
