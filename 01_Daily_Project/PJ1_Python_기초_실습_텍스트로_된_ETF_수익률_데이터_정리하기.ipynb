```python
from statistics import mean # mean 함수 사용을 위해 import
```

1. input()을 통해 ETF 종목(티커)들과 수익률 리스트를 받아야 함.
    1. 종목 티커를 먼저 input으로 받는다.
    2. 받은 문자열을 어떻게 구분할 지 split()
      - 구분자는 공백(" "), 쉼표(","), ("#")만 존재하는 것으로 가정.
    3. 수익률은 소수점으로 이루어져있기 때문에 구분 후 float형으로 변환까지 필요
2. 대소문자 정리
    1. 문자열로 받아 구분하여 리스트로 만든 종목 리스트를 모두 대문자로 변환(Ticker)
3. 계산 가능하게 자료형을 숫자로 저장

```python
ticker=input("티커를 입력해주세요: ") # ETF Ticker를 입력받아
ticker=ticker.replace(","," ").replace("#"," ").split() # 구분자를 공백으로 정리하고 리스트로 변환
ticker_up=[x.upper() for x in ticker] # List Comprehension을 활용한 대문자 전환.
ticker_list=list(ticker_up) # 문자열 모음을 리스트로 변환
print(ticker_list)
```

```python
ror=input("수익률을 입력해주세요: ") # ETF 수익률 입력받아
ror=ror.replace(","," ").replace("#"," ").split() # 구분자를 공백으로 정리하고 리스트로 변환
ror_list=list(ror) # RoR 문자열을 리스트로 만들어줌.
print(ror_list)
ror_float=[float(x) for x in ror_list] #List Comprehension을 통해 ror리스트를 실수형으로 변환
print(ror_float)

```

4. 개수가 맞는지 판별
    1. len()==len()
5. 수익률의 평균을 계산해야 함. mean() or sum()/len()
    - from statistics import mean()
6. 수익률이 평균보다 높은 경우를 선별
    1. for 구문의 if 조건문 x > mean # 초과

```python
ticker_len=len(ticker_list) # 개수 판별을 위해 ticker_len 구하기
print(ticker_len)
ror_len=len(ror_float)
print(ror_len)
if ticker_len!=ror_len: # Error Clause : 오류 발생 경우를 먼저 쳐내고
  print("종목 수와 수익률 개수는 같아야 합니다!")
else:
  ror_mean=mean(ror_float) # ror float 리스트의 평균을 구함.
  print(ror_mean)
  good_ror={}
  # ror이 평균보다 높은 경우를 뽑아 새로운 리스트로 만들어야 함... 근데 두 리스트를 쌍으로 만들어야 되는데 어떻게 하지? => zip
  for t, r in zip(ticker_list,ror_float):
    if r>ror_mean:
      good={t:r} # Ticker와 RoR을 묶기 위해 딕셔너리가 필요했음. 이전에 한 번 들어본 적은 있지만 써본 적은 처음이라 생소했다.
      good_ror[t]=r # good_ror 딕셔너리에 t(key)와 r(value) 한 쌍을 추가한다.
                    # good_ror.update(good)도 가능! update는 두 개의 dictionary를 합칠 때 사용
  print(good_ror)   # 결과 확인!!!!!!!!!!!
                    # 20260106 / PM 08:47 일단 여기까지 마무리하고, ticker 입력값이 문자열이 아닌 경우, ror 입력값이 실수형이 아닌 경우에
                    # try/except문으로 ValueError 회피하는 것도 추가해볼 수 있을 것 같다.
```
