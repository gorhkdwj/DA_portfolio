Step 1: 데이터 초기화 (Mock Data Setup)
API에서 막 데이터를 받아왔다고 가정하고, 아래와 같은 구조로 portfolio 리스트를 만드세요.

portfolio = [
    {"ticker": "SPY", "ror": 10.5},
    {"ticker": "QQQ", "ror": 15.2},
    {"ticker": "TLT", "ror": -1.5}
]


```python
portfolio = [
    {"ticker": "SPY", "ror": 10.5},
    {"ticker": "QQQ", "ror": 15.2},
    {"ticker": "TLT", "ror": -1.5}
]
print("1. 초기 데이터: ")
for item in portfolio:
    print(f"{item['ticker']}: {item['ror']}")
```

Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
상황: 새로운 데이터 {"ticker": "SPY", "ror": 11.2}와 {"ticker": "NVDA", "ror": 25.0}이 들어왔습니다.

미션:

SPY는 이미 있으니 수익률을 11.2로 업데이트해야 합니다.

NVDA는 없으니 리스트에 **추가(append)**해야 합니다.

이를 처리하는 for 반복문 로직을 작성하세요. (힌트: found 플래그 변수 활용)

```python
# Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
# [설명] 사용자로부터 티커를 입력받아, 기존에 있으면 수정하고 없으면 추가하는 로직입니다.

while True: # 1 조건 반복문: 사용자가 종료를 원할 때까지 계속 프로그램을 실행합니다.
    try:
        # 2 사용자 입력: 대문자로 변환
        target_ticker = input("추가 또는 수정할 티커를 입력하세요.(종료는 'exit' 입력): ").upper()
    except ValueError: 
        # 예외 처리: 입력 과정에서 문자가 아닌 값이 입력될 경우
        print("티커는 문자로 입력해주세요.")
        continue

    if target_ticker == 'EXIT': # 3 종료 조건: 'EXIT'를 입력하면 프로그램을 멈춥니다(break).
        print("입력을 종료합니다.")
        break

    # 4 Upsert Logic: found 변수 플래그를 통해 기존 포트폴리오와 비교합니다.
    # 처음엔 못 찾았다고 가정(False)하고, 리스트를 뒤져서 찾으면 True로 바꿉니다.
    found = False

    for item in portfolio: # 포트폴리오(리스트)에 있는 item들을 하나씩 꺼내서 확인
        if item['ticker'] == target_ticker: # 5 기존에 티커가 있는 경우
            found = True # 깃발을 세웁니다(True).

            # 6 수정 모드: 이미 존재하므로 수익률만 새로 입력받아 업데이트합니다.
            while True:
                try:
                    fix_ror = float(input("수정할 포트폴리오의 수익률을 입력하세요.(종료는 'exit' 입력): "))
                    item['ror'] = fix_ror # 해당 아이템의 'ror' 값을 덮어씁니다 (Update)
                    print(f"수정한 포트폴리오: 티커={item['ticker']} / 수익률={item['ror']}")
                    break # 수정이 끝나면 수정 모드 반복문 탈출
                except ValueError:
                    print("수익률은 숫자로 입력해주세요.")
                    continue
            break # 이미 찾아서 수정했으므로, 더 이상 포트폴리오를 뒤질 필요가 없습니다. 탐색 종료, 티커 input부터 다시 시작.

    if found == False: # 7 포트폴리오(리스트)를 끝까지 다 뒤는데 못 찾은 경우 (신규 추가)
        try:
            # 새로운 수익률 정보를 입력받습니다.
            new_ror = float(input("추가할 포트폴리오의 수익률을 입력하세요.(종료는 'exit' 입력): "))
            
            # 8 추가(Append): 리스트의 맨 끝에 새로운 딕셔너리를 만들어 넣습니다.
            portfolio.append({'ticker': target_ticker, 'ror': new_ror})
            print(f"추가한 포트폴리오: 티커={target_ticker} / 수익률={new_ror}")
        except ValueError:
            print("수익률은 숫자로 입력해주세요.")
            continue

# 9 결과 확인: 최종적으로 변경된 포트폴리오 상태를 출력합니다.
print("추가 및 수정이 완료된 포트폴리오: ")
for item in portfolio: # for문을 활용해 보기 좋게 출력
    print(f"티커={item['ticker']} / 수익률={item['ror']}")
```

Step 3: 데이터 삭제 (Safe Removal)
상황: "채권(TLT)은 포트폴리오에서 뺍시다."

미션: 리스트 내 딕셔너리 중 "ticker"가 "TLT"인 항목을 찾아 삭제하세요.

주의: 리스트를 for문으로 돌면서 삭제하면 인덱스 에러가 날 수 있습니다. 삭제할 대상을 먼저 찾고(target), 나중에 remove(target) 하는 방식을 고민해 보세요.

```python
# 반복문으로 돌면서 삭제할 티커가 기존 티커에 있는지 확인
# 맞는 티커가 있다면 해당 티커(딕셔너리)를 리스트에서 삭제
# 단, for문으로 돌면서 삭제하면 인덱스 에러가 날 수 있다.



while True:
    try:
        del_ticker = input("삭제할 포트폴리오의 티커를 입력하세요.(종료는 'exit' 입력): ").upper()
    except ValueError:
        print("티커는 문자로 입력해주세요.") # 문자 외 입력시 ValueError
        continue
    if del_ticker == 'EXIT': # exit 입력시 종료
        print("입력을 종료합니다.") 
        break
    # for 문을 통해 삭제할 대상에 마킹해두고, 나중에 제거한다! -> enumerate 사용(del_ticker==ticker -> i 마킹!)
    found = False
    for i, item in enumerate(portfolio):
        if item['ticker'] == del_ticker:
            found = True
            del portfolio[i]
            print(f"삭제한 포트폴리오: 티커={del_ticker}")
            break
    if not found: # 리스트 다 돌았는데 못찾았다!
        print(f"티커 {del_ticker}를 찾을 수 없습니다.") # 다시 While문 부터 반복
print("삭제가 완료된 포트폴리오: ")
for item in portfolio:
    print(f"티커={item['ticker']} / 수익률={item['ror']}")
```

Step 4: 인덱싱 및 조회 (Indexing & Search)
상황: "데이터가 10만 개면 for문 돌다가 서버 다운됩니다. 티커만 치면 바로 나오게 해주세요."

미션: portfolio 리스트를 { "SPY": 11.2, "QQQ": 15.2, ... } 형태의 딕셔너리(market_map)로 변환하세요. (Dictionary Comprehension 활용)

테스트: 변환된 딕셔너리에서 NVDA의 수익률을 O(1) 속도로 조회해 보세요.

```python
# 전산학의 Big O 표기에 따르면, List는 O(n) // Dictionary는 O(1) 라는 차이를 갖고 있다.
# List는 포함한 데이터가 10만개면 최악의 경우 10만번 뒤져야 하지만, Dictionary는 key만 알고 있으면 단 한번에 원하는 값을 찾을 수 있다.
# 그에 따라, 컴퓨팅 리소스를 줄이기 위해, Step 4.의 요구사항이 발생하는 것이다.
# 그리고 그 방법으로 Dictionary Comprehension을 활용할 것을 요구하는 중이다!

market_map={item['ticker']:item['ror'] for item in portfolio}

# test
print(f"NVDA의 수익률: {market_map['NVDA']}")
```
