Step 1: 데이터 초기화 (Mock Data Setup)
API에서 막 데이터를 받아왔다고 가정하고, 아래와 같은 구조로 portfolio 리스트를 만드세요.

portfolio = [
    {"ticker": "SPY", "ror": 10.5},
    {"ticker": "QQQ", "ror": 15.2},
    {"ticker": "TLT", "ror": -1.5}
]


```python
portfolio = [
    {"ticker": "SPY", "ror": 10.5},
    {"ticker": "QQQ", "ror": 15.2},
    {"ticker": "TLT", "ror": -1.5}
]
print("1. 초기 데이터: ")
for item in portfolio:
    print(f"{item['ticker']}: {item['ror']}")
```

Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
상황: 새로운 데이터 {"ticker": "SPY", "ror": 11.2}와 {"ticker": "NVDA", "ror": 25.0}이 들어왔습니다.

미션:

SPY는 이미 있으니 수익률을 11.2로 업데이트해야 합니다.

NVDA는 없으니 리스트에 **추가(append)**해야 합니다.

이를 처리하는 for 반복문 로직을 작성하세요. (힌트: found 플래그 변수 활용)

```python
# Step 2: 데이터 갱신 로직 (The "Upsert" Logic)
# [설명] 사용자로부터 티커를 입력받아, 기존에 있으면 수정하고 없으면 추가하는 로직입니다.

while True: # [1] 무한 반복: 사용자가 종료를 원할 때까지 계속 프로그램을 실행합니다.
    try:
        # [2] 사용자 입력: 대문자로 변환(.upper())하여 처리 (사용자가 spy라고 입력해도 SPY로 인식)
        target_ticker = input("추가 또는 수정할 티커를 입력하세요.(종료는 'exit' 입력): ").upper()
    except ValueError: 
        # [예외 처리] 입력 과정에서 예상치 못한 에러가 발생하면 넘어갑니다.
        print("티커는 문자로 입력해주세요.")
        continue

    if target_ticker == 'EXIT': # [3] 종료 조건: 'EXIT'를 입력하면 프로그램을 멈춥니다(break).
        print("입력을 종료합니다.")
        break

    # [4] 탐색 시작: 'found' 변수를 깃발(Flag)처럼 사용합니다.
    # 처음엔 못 찾았다고 가정(False)하고, 리스트를 뒤져서 찾으면 True로 바꿉니다.
    found = False

    for item in portfolio: # 포트폴리오(리스트)에 있는 아이템들을 하나씩 꺼내서 확인
        if item['ticker'] == target_ticker: # [5] 찾았다! (기존에 티커가 있는 경우)
            found = True # 깃발을 세웁니다 (찾았음 표시)
            
            # [6] 수정 모드: 이미 존재하므로 수익률만 새로 입력받아 업데이트합니다.
            while True:
                try:
                    fix_ror = float(input("수정할 포트폴리오의 수익률을 입력하세요.(종료는 'exit' 입력): "))
                    item['ror'] = fix_ror # 해당 아이템의 'ror' 값을 덮어씁니다 (Update)
                    print(f"수정한 포트폴리오: 티커={item['ticker']} / 수익률={item['ror']}")
                    break # 수정이 끝나면 수정 모드 반복문 탈출
                except ValueError:
                    print("수익률은 숫자로 입력해주세요.")
                    continue
            break # 이미 찾아서 수정했으므로, 더 이상 포트폴리오를 뒤질 필요가 없습니다. 탐색 종료.

    if found == False: # [7] 끝까지 다 뒤는데 못 찾은 경우 (신규 추가)
        try:
            # 새로운 수익률 정보를 입력받습니다.
            new_ror = float(input("추가할 포트폴리오의 수익률을 입력하세요.(종료는 'exit' 입력): "))
            
            # [8] 추가(Append): 리스트의 맨 끝에 새로운 딕셔너리를 만들어 넣습니다.
            portfolio.append({'ticker': target_ticker, 'ror': new_ror})
            print(f"추가한 포트폴리오: 티커={target_ticker} / 수익률={new_ror}")
        except ValueError:
            print("수익률은 숫자로 입력해주세요.")
            continue

# [9] 결과 확인: 최종적으로 변경된 포트폴리오 상태를 출력합니다.
print("추가 및 수정이 완료된 포트폴리오: ")
for item in portfolio:
    print(f"티커={item['ticker']} / 수익률={item['ror']}")
```
